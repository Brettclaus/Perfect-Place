```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import xlrd
```

```python
excel_file_path = "Crime_Rates_by_Metropolitan_Areas_2016.xls"

# Read only the third sheet named "metros alphabetical", so that we are including all metros in order 
crime_rates_df = pd.read_excel(excel_file_path, sheet_name="Metros Alphabetical")

# Display the DataFrame
crime_rates_df
```

```python
print(crime_rates_df.columns)
```

```python
# Extract City and State information from the first column for readability
crime_rates_df[['City', 'State']] = crime_rates_df.iloc[:, 0].str.extract(r'([^\d]+), (\w{2})')

# Remove 'M.S.A.' from 'City' column
crime_rates_df['City'] = crime_rates_df['City'].str.replace(' M.S.A.', '')

# Display the updated DataFrame
crime_rates_df
```

```python
print(crime_rates_df.columns)
```

```python
# Rename columns for readability
crime_rates_df = crime_rates_df.rename(columns={'Note that not all areas show Violent Crime Totals.':'Rate',
    'Violent\ncrime': 'Violent Crime',
    'Murder and\nnonnegligent\nmanslaughter': 'Murder and Non Negligent Manslaughter',
    'Rape1': 'Rape',
    'Aggravated\nassault': 'Aggravated Assault',
    'Property\ncrime': 'Property Crime',
    'Larceny-\ntheft': 'Larceny and Theft',
    'Motor\nvehicle\ntheft': 'Motor Vehicle Theft'
})

# Rearrange columns
crime_rates_df = crime_rates_df[['City', 'State', 'Population','Rate', 'Violent Crime', 'Murder and Non Negligent Manslaughter', 'Rape', 'Robbery', 'Aggravated Assault', 'Property Crime', 'Burglary', 'Larceny and Theft', 'Motor Vehicle Theft']]

# Display the updated DataFrame
crime_rates_df
```

```python
# Remove rows where 'City' column is blank
crime_rates_df = crime_rates_df.dropna(subset=['City'])

# Display the updated DataFrame
crime_rates_df
```

```python

# Load the CSV file, doing so first avoids warning or potential issues going forward when changing data frame
crime_rates_df = pd.read_csv('crime_rates_df.csv')  

# Calculate the actual number of each crime
crime_rates_df['Actual Violent Crime'] = (crime_rates_df['Violent Crime'] * crime_rates_df['Population']) / 100000
crime_rates_df['Actual Murder and Non Negligent Manslaughter'] = (crime_rates_df['Murder and Non Negligent Manslaughter'] * crime_rates_df['Population']) / 100000
crime_rates_df['Actual Rape'] = (crime_rates_df['Rape'] * crime_rates_df['Population']) / 100000
crime_rates_df['Actual Robbery'] = (crime_rates_df['Robbery'] * crime_rates_df['Population']) / 100000
crime_rates_df['Actual Aggravated Assault'] = (crime_rates_df['Aggravated Assault'] * crime_rates_df['Population']) / 100000
crime_rates_df['Actual Property Crime'] = (crime_rates_df['Property Crime'] * crime_rates_df['Population']) / 100000
crime_rates_df['Actual Burglary'] = (crime_rates_df['Burglary'] * crime_rates_df['Population']) / 100000
crime_rates_df['Actual Larceny and Theft'] = (crime_rates_df['Larceny and Theft'] * crime_rates_df['Population']) / 100000
crime_rates_df['Actual Motor Vehicle Theft'] = (crime_rates_df['Motor Vehicle Theft'] * crime_rates_df['Population']) / 100000

# Calculate the percentage of each crime
crime_rates_df['Percentage Violent Crime'] = (crime_rates_df['Actual Violent Crime'] / crime_rates_df['Population']) * 100
crime_rates_df['Percentage Murder and Non Negligent Manslaughter'] = (crime_rates_df['Actual Murder and Non Negligent Manslaughter'] / crime_rates_df['Population']) * 100
crime_rates_df['Percentage Rape'] = (crime_rates_df['Actual Rape'] / crime_rates_df['Population']) * 100
crime_rates_df['Percentage Robbery'] = (crime_rates_df['Actual Robbery'] / crime_rates_df['Population']) * 100
crime_rates_df['Percentage Aggravated Assault'] = (crime_rates_df['Actual Aggravated Assault'] / crime_rates_df['Population']) * 100
crime_rates_df['Percentage Property Crime'] = (crime_rates_df['Actual Property Crime'] / crime_rates_df['Population']) * 100
crime_rates_df['Percentage Burglary'] = (crime_rates_df['Actual Burglary'] / crime_rates_df['Population']) * 100
crime_rates_df['Percentage Larceny and Theft'] = (crime_rates_df['Actual Larceny and Theft'] / crime_rates_df['Population']) * 100
crime_rates_df['Percentage Motor Vehicle Theft'] = (crime_rates_df['Actual Motor Vehicle Theft'] / crime_rates_df['Population']) * 100

# print CSV for map layer
crime_rates_df.to_csv('crime_rates_df.csv', index=False)  

```

```python
# Select the columns for the graph to visualize Percentage of crimes and population
selected_columns = ['Population', 'Percentage Violent Crime', 'Percentage Murder and Non Negligent Manslaughter',
                    'Percentage Rape', 'Percentage Robbery', 'Percentage Aggravated Assault',
                    'Percentage Property Crime', 'Percentage Burglary', 'Percentage Larceny and Theft',
                    'Percentage Motor Vehicle Theft']

# Gathering Standard Deviation within each state for each city to better visualize the different categories of crime across the U.S.
# Group by 'State' and calculate mean and standard deviation for each crime category 
state_crime_summary = crime_rates_df.groupby('State')[selected_columns].agg(['mean', 'std', 'sum']).reset_index()

# Plotting for each crime category, excluding population
for crime_category in selected_columns[1:]:  
    plt.figure(figsize=(12, 6))
    sns.barplot(x='State', y=(crime_category, 'mean'), data=state_crime_summary, errorbar='sd', capsize=0.2, palette='viridis')
    plt.title(f'Mean {crime_category} Percentage for Each City within Each State')
    plt.xlabel('State')
    plt.ylabel(f'Mean {crime_category} Percentage')
    plt.xticks(rotation=45, ha='right')  

# Show all the plots
plt.show()
```
