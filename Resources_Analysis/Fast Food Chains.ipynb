# Fast Food Restaurants Across America

"This is a list of 10,000 fast food restaurants provided by Datafiniti's Business Database. The dataset includes the restaurant's address, city, latitude and longitude coordinates, name, and more."

```python
# Import necessary library
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Fast_Food_US.csv')  

# Group the data by 'city' and count the number of fast food chains in each city
city_counts = df['city'].value_counts().reset_index()
city_counts.columns = ['City', 'NumberOfChains']

# Display the cities with the most fast food chains
city_counts.head() 

# If you want to save the result to a CSV file
city_counts.to_csv('fast_food_chains_by_city.csv', index=False)
```

```python

# Group the data by 'city' and count the number of fast food chains in each city
city_counts = df['city'].value_counts().reset_index()
city_counts.columns = ['City', 'NumberOfChains']

# Get the top 10 cities with the most fast food chains
top_cities = city_counts.head(10)

# Create a bar chart
plt.figure(figsize=(12,8))
plt.bar(top_cities['City'], top_cities['NumberOfChains'], color='skyblue')
plt.title('Top 10 Cities with the Most Fast Food Chains')
plt.xlabel('City')
plt.ylabel('Number of Fast Food Chains')
plt.xticks(rotation=45)
plt.tight_layout() 

# Display the bar chart
plt.show()
```

```python

# Filter for 'Taco Bell' in the 'name' column
taco_bell_df = df[df['name'].str.contains('Taco Bell', case=False, na=False)]

# Group by city and count the number of Taco Bell locations
taco_bell_city_counts = taco_bell_df.groupby('city').size().reset_index(name='NumberOfTacoBellLocations')

# Sort the cities by the number of Taco Bell locations and get the top 10
top_taco_bell_cities = taco_bell_city_counts.sort_values(by='NumberOfTacoBellLocations', ascending=False).head(10)

# Plotting the results in a bar chart
plt.figure(figsize=(12, 8))
plt.bar(top_taco_bell_cities['city'], top_taco_bell_cities['NumberOfTacoBellLocations'], color='blue')
plt.title('Top 10 Cities with the Most Taco Bell Locations')
plt.xlabel('City')
plt.ylabel('Number of Taco Bell Locations')
plt.xticks(rotation=45)
plt.tight_layout() 
```

```python

# Filter for 'McDonald's' in the 'name' column
mcdonalds_df = df[df['name'].str.contains("McDonald's", case=False, na=False)]

# Group by city and count the number of McDonald's locations
mcdonalds_city_counts = mcdonalds_df.groupby('city').size().reset_index(name='NumberOfMcDonaldsLocations')

# Sort the cities by the number of McDonald's locations and get the top 10
top_mcdonalds_cities = mcdonalds_city_counts.sort_values(by='NumberOfMcDonaldsLocations', ascending=False).head(10)

# Plotting the results in a bar chart
plt.figure(figsize=(12, 8))
plt.bar(top_mcdonalds_cities['city'], top_mcdonalds_cities['NumberOfMcDonaldsLocations'], color='red')
plt.title('Top 10 Cities with the Most McDonald\'s Locations')
plt.xlabel('City')
plt.ylabel('Number of McDonald\'s Locations')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust the plot to ensure everything fits without overlapping
plt.show()
```

```python

# Load the dataset
df = pd.read_csv('Fast_Food_US.csv')  # Replace with the actual path to your CSV file

# Grouping data by city and counting unique chains
city_diversity = df.groupby('city')['name'].nunique().reset_index()
city_diversity.columns = ['City', 'UniqueFastFoodChains']

# Sorting the cities by the number of unique chains and selecting the top 20
top_20_diverse_cities = city_diversity.sort_values(by='UniqueFastFoodChains', ascending=False).head(20)

top_20_diverse_cities
```

```python

# Group by state and count the number of restaurants
state_counts = df['province'].value_counts().reset_index()
state_counts.columns = ['State', 'NumberOfRestaurants']

# Getting the top 3 states with the most restaurants
top_3_states_most = state_counts.head(3)

# Getting the top 3 states with the least restaurants
top_3_states_least = state_counts.tail(3)

# Plotting the top 3 states with the most fast food restaurants
plt.figure(figsize=(10, 5))
plt.bar(top_3_states_most['State'], top_3_states_most['NumberOfRestaurants'], color='green')
plt.title('Top 3 States with Most Fast Food Restaurants')
plt.xlabel('State')
plt.ylabel('Number of Restaurants')
plt.show()

# Plotting the top 3 states with the least fast food restaurants
plt.figure(figsize=(10, 5))
plt.bar(top_3_states_least['State'], top_3_states_least['NumberOfRestaurants'], color='red')
plt.title('Top 3 States with Least Fast Food Restaurants')
plt.xlabel('State')
plt.ylabel('Number of Restaurants')
plt.show()
```
