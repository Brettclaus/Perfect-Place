```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import xlrd
```

```python
#Read in excel file for Walk, Bike, Transit Score
walk_bike_transit_df = pd.read_excel("Walk_Bike_Transit_Score.xlsx")

# Display the DataFrame
walk_bike_transit_df
```

```python
# Bar graph for top 10 cities with the best Bike scores
sns.barplot(x='City', y='Bike Score', data=walk_bike_transit_df.nlargest(10, 'Bike Score'))
plt.title('Top 10 Cities with Best Bike Scores')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Replacing the cities with no transit score ('--') with NAN in order to continue with analysis 
walk_bike_transit_df['Transit Score'] = pd.to_numeric(walk_bike_transit_df['Transit Score'], errors='coerce')

# Bar graph for top 10 cities with the best Transit scores
sns.barplot(x='City', y='Transit Score', data=walk_bike_transit_df.nlargest(10, 'Transit Score'))
plt.title('Top 10 Cities with Best Transit Scores')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Bar graph for top 10 cities with the best Walking scores
sns.barplot(x='City', y='Walk Score', data=walk_bike_transit_df.nlargest(10, 'Walk Score'))
plt.title('Top 10 Cities with Best Walking Scores')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Create Average Travel Score Column
walk_bike_transit_df['Average Travel Score'] = walk_bike_transit_df[['Bike Score', 'Walk Score', 'Transit Score']].mean(axis=1)

#Reorganize Data Frame and Create Bar Graph for Top 10 Average Travel Scores
top_10_average_scores_df = walk_bike_transit_df.nlargest(10, 'Average Travel Score')

top_10_average_scores_df
```

```python
# Bar graph for top 10 cities with the highest Average Travel Scores
sns.barplot(x='City', y='Average Travel Score', data=top_10_average_scores_df)
plt.title('Top 10 Cities with Highest Average Travel Scores')
plt.xticks(rotation=45,ha='right')
plt.show()
```

```python
top_10_average_scores_df.to_csv("top_10_average_Travel_Scores", index=False)
```

```python
sns.pairplot(walk_bike_transit_df, vars=['Walk Score', 'Transit Score', 'Bike Score', 'Average Travel Score'], diag_kind='kde')
plt.suptitle('Pairplot of Scores and Average Travel Score', y=1.02)
plt.show()
```

Initially examining this pairplot, there seems to be a positive correlation between a majority of the relationships of Average Travel Score and each of the different categories. Below, I am further exploring the individual relationships.  

P-Value for how heavily related the Walk Score is to the overall Transi

```python
# importing pearsonr in order to run tests for relationships between the Average Travel Score and Walk, Bike, and Transit
from scipy.stats import pearsonr

# Example: Calculate correlation coefficient and p-value for Walk Score and Average Travel Score
correlation_coefficient, p_value = pearsonr(walk_bike_transit_df["Walk Score"], walk_bike_transit_df["Average Travel Score"])

print("Correlation Coefficient:", correlation_coefficient)
print("P-value:", p_value)
```

```python
correlation_coefficient, p_value = pearsonr(walk_bike_transit_df["Bike Score"], walk_bike_transit_df["Average Travel Score"])

print("Correlation Coefficient:", correlation_coefficient)
print("P-value:", p_value)
```

```python
#First dropping any NAN's in Transit score
walk_bike_transit_df.dropna(subset=["Transit Score"], inplace=True)

correlation_coefficient, p_value = pearsonr(walk_bike_transit_df["Transit Score"], walk_bike_transit_df["Average Travel Score"])

print("Correlation Coefficient:", correlation_coefficient)
print("P-value:", p_value)
```
